```kotlin
val favoriteFruits = listOf("Mango", "Blueberry", "Pineapple", "Watermelon", "Kiwi")
println("List of fruits in Uppercase:")
for (favoriteFruit in favoriteFruits) {
    println(favoriteFruit.uppercase())
}
println("\nTotal number of fruits: ${favoriteFruits.size}")
```

```kotlin
var score =90
when  {
    score == 100 -> println("Perfect score!")
    score < 0 ||  score > 100 -> println("Invalid score")
    score in 0..49 -> println("You failed!")
    score in 50..60 -> println("Just passed!")
    else -> println("Well done!")
}
```
